# Step 1- load libraries

```python
import pandas as pd
```

```python
import yfinance as yf
```

```python
import datetime
from datetime import timedelta, date
```

```python
import plotly.graph_objects as go
```

```python
import plotly.express as px
```

# step 2- get d1, d2

```python
d1 = date.today().strftime("%Y-%m-%d")
```

```python
type(d1)
```

```python
d1
```

```python
d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
```

```python
d2
```

# step 3- Get data

```python
data = yf.download('META', 
                      start=d2, 
                      end=d1, 
                      progress=True)
```

```python
data
```

```python
data.reset_index(inplace=True)
```

```python
data
```

# Step 4- Visualise

# Candlestick

```python
data2= go.Candlestick(x=data["Date"],open=data["Open"], high=data["High"],  low=data["Low"], close=data["Close"])
data2
```

```python
figure = go.Figure(data=data2)
figure.update_layout(title = "Meta Stock Price Analysis", xaxis_rangeslider_visible=False)
```

# Rangeslider

```python
figure = px.line(data, x='Date', y='Close', 
                 title='Stock Market Analysis with Rangeslider')
figure.update_xaxes(rangeslider_visible=True)
figure.show()
```
